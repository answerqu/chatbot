# Чат-бот
#### Общая идея:
Наш чат-бот распознает 5 запросов, это: запрос времени, погоды в Новосибирске, списка умений чат-бота, книг по програмированию, книг по машинному обучению. По собственноручно созданному датасету из простых команд обучалась модель XGBClassifier (n_estimators=10, booster='gblinear'). Каждый запрос в виде строки предварительно очищался от знаков пунктуации, местоимений, предлогов, союзов, частиц, все слова приводились к начальной форме. Затем каждый запрос кодировался one-hot методом в массив размера 144 (количество получившихся уникальных слов). То есть на i-ой позиции стоит значение 1, если i-ое слово присутствует в этом предложении.

Если в запросе пользователя есть ключевые слова (слова благодарности, прощания или приветствия), то сначала чат-бот отвечает на них. Затем, если запрос содержит ключевую тему, бот выдает ответ по запросу. 

#### Python scripts - описание:

texts.py - содержит списки текстовых запросов, на основе которых происходило обучение модели, списки ответов и словарь уникальных слов.

functions.py - основные функции (функции ответов, функции предобработки предложений и работы с обученной моделью)

model.py - скрипт с подготовкой массива на обучение и с самим обучением модели

bot.py - содержит одну единственную функцию - запуск чат-бота
